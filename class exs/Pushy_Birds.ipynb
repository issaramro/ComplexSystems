```python
import numpy as np 
import math
import matplotlib.pyplot as plt 
```

```python
density_of_birds = []
ar = np.zeros(100)
nb_of_birds = 0

for i in range(10000):

    n = np.random.randint(0,100)
    
    if n != 0 and ar[n-1] == 1:
        ar[n-1] = 0
        ar[n] = 1

    elif n != 99 and ar[n+1] == 1:
        ar[n+1] = 0
        ar[n] = 1
        
    elif n != 0 and n != 99 and ar[n-1] == 1 and ar[n+1] == 1:
        ar[n-1] = 0
        ar[n+1] = 0
        ar[n] = 1
    
    else:
        ar[n] = 1
    
    nb_of_birds = np.sum(ar/100)
    density_of_birds.append(nb_of_birds)

plt.scatter(np.arange(len(density_of_birds)), density_of_birds)
plt.plot(np.arange(len(density_of_birds)), 0.33*np.ones(len(density_of_birds)), color = 'r')
plt.show()
print(density_of_birds[-1])
```

```python
def F(k):
    return 2**(k+1)*(k+1)/math.factorial(k+3)

V_arr = []
for k in range(100):
    V = F(k) - F(k+1)
    V_arr.append(V)

plt.scatter(np.arange(0,100), V_arr)
print(V_arr)
```
